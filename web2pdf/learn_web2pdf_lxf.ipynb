```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
_content = requests.get('http://www.liaoxuefeng.com/wiki/0013739516305929606dd18361248578c67b8067c8c017b000')
```

```python
soup = bs(_content.content, "html.parser")
soup
```

### parse menu

```python
menu_tag = soup.find_all(class_="uk-nav uk-nav-side")[1]
soup.find_all(class_="uk-nav uk-nav-side")
```

```python
menu_tag
```

```python
def parseMenu():
    for li in menu_tag.find_all("li"):
        url = li.a.get("href")
        if not url.startswith("http"):
            url = "".join(['www.liaoxuefeng.com', url])  # 补全为全路径
        yield url
```

```python
## 1.interator
for a in  parseMenu():
    print (a)
```

```python
##2.next   python3.* next->__next__()
f = parseMenu()
```

```python
f.__next__()
```

### parse body

```python
body = soup.find_all(class_="x-wiki-content")[0]
soup.find_all(class_="x-wiki-content")
```

```python
title = soup.find('h4').get_text()
title
```

```python
center_tag.insert(1, title_tag)
body.insert(1, center_tag)
```

```python
import re
html = str(body)
# body中的img标签的src相对路径的改成绝对路径
pattern = "(<img .*?src=\")(.*?)(\")"

def func(m):
    if not m.group(3).startswith("http"):
        rtn = "".join([m.group(1), 'www.liaoxuefeng.com', m.group(2), m.group(3)])
        return rtn
    else:
        return "".join([m.group(1), m.group(2), m.group(3)])

html = re.compile(pattern).sub(func, html)
```

```python
html
```

```python
html_template = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
</head>
<body>
{content}
</body>
</html>
"""
```

```python
html = html_template.format(content=html)
html = html.encode("utf-8")
```
